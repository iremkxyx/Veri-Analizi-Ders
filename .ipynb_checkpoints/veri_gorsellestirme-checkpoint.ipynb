### bar grafikleri

```python
# plt.plot değil plt.bar
liste1 = ["X","Y","Z","T"]
liste2 = [10,20,15,5]
plt.bar(liste1,liste2)
plt.title("Başlık")
plt.show() # ek bilgi görmemek için
```

```python
liste1 = ["X","Y","Z","T"]
liste2 = [10,20,15,5]
plt.bar(liste1,liste2,width=0.5,color="red")
plt.title("Başlık")
plt.show()
```

```python
liste1 = ["X","Y","Z","T"]
liste2 = [10,20,15,5]
plt.barh(liste1,liste2,height=0.5,color="red")
plt.title("Başlık")
plt.show()
```

```python
liste1 = ["X","Y","Z","T"]
liste2 = [10,20,15,5]
liste3 = [20,40,30,10]
plt.bar(liste1,liste3,width=0.5,color="black")
plt.bar(liste1,liste2,width=0.3,color="red")

plt.title("Başlık")
plt.xlabel("Yatay eksen")
plt.ylabel("Düşey eksen")
plt.legend(labels= ["Siyah","Kırmızı"])
plt.show()
```

```python
liste1 = [10,20,50,100]
liste2 = [5,8,9,50]
liste3 = ["A","B","C","D"]
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,5)) #2,1 alt alta 1,2 yan yana grafik
plt.bar(liste3,liste1, color="blue")
plt.bar(liste3,liste2,color="orange")
plt.title("Başlık",size =18)
plt.xlabel("Düşey",size=15)
plt.ylabel("Yatay",size=15)
plt.legend(labels=["Mavi","Turuncu"])
fig.text(0.85,0.7,
         """ 
         Bu bölümde grafik ile ilgili
         açıklamalar,ek bilgiler 
         bulunmaktadır.
         """, fontsize= 15)
plt.show()
```

### dağılım grafikleri

```python
a = (np.random.rand(15,10)*100).round()
b = (np.random.randn(15,10)*100).round()
```

```python
plt.scatter(a,b)
```

```python
x = (np.random.rand(15,10)*100).round()
y = (np.random.rand(15,10)*100).round()

x2 = (np.random.rand(15,10)*50).round()
y2 = (np.random.rand(15,10)*50).round()

```

```python
plt.scatter(x,y)
plt.scatter(x2,y2)
```

```python
plt.scatter(x,y,color="black", alpha=0.5) #alpha saydamlık
plt.scatter(x2,y2,color="cyan")
```

```python
plt.scatter(x,y,color="black", alpha=0.5)
plt.scatter(x2,y2,color="red")
plt.xlabel("Yatay eksen", size=15)
plt.ylabel("Düşey eksen", size=15)
plt.title("Grafiğin başlığı",size=15)
```

```python
fig, ax= plt.subplots(1,1,figsize=(5,5))
plt.scatter(x,y,color="black", alpha=0.5)
plt.scatter(x2,y2,color="red")
plt.xlabel("Yatay eksen", size=15)
plt.ylabel("Düşey eksen", size=15)
plt.title("Grafiğin başlığı",size=15)
plt.legend(labels=["siyah","kırmızı"])
```

#figure ün içinden fig.text diye açıklama oluşturabiliriz

```python
x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])

renk=["black","blue","brown","red","green"]

boyut= [100,200,400,800,1600]

plt.scatter(x,y,color=renk,s=boyut)
plt.show()
```

### 3d grafikler

```python
from mpl_toolkits import mplot3d
```

```python
fig=plt.figure(figsize=(8,8))
ax= plt.axes(projection="3d")

z= np.random.rand(12,4)*10
x= z*np.random.rand(12,4)
y= z*np.random.rand(12,4)

ax.plot_surface(x,y,z,color= "r")
plt.title("3 Boyutlu Grafik")
plt.show()
```
