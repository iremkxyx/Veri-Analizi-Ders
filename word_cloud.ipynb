```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
oku = open("metin.txt", encoding="utf-8").read()
oku
```

```python
kelimebulutu= WordCloud().generate(oku)
plt.imshow(kelimebulutu)
plt.show()
```

```python
plt.figure(figsize=(8,6))
kelimebulutu= WordCloud(background_color= "white",
                       width= 1200,
                       height= 1000).generate(oku)
plt.imshow(kelimebulutu)
plt.show()
```

```python
text="""
Ana düşünce (Ishmael için denizin, maneviyatın bir metaforu
olduğu düşüncesi) burada paragrafa öncülük etmektedir, 
fakat bu düşünce verilmeden önce genellikle bir geçiş ifadesi 
veya cümlesi gelir. Daha farklı paragraf türlerini görmek için
açıklamalı deneme örneğimizi inceleyebilirsiniz.
"""
text
```

```python
gorsel= WordCloud(background_color= "white").generate(text)
plt.imshow(gorsel)
plt.show()
```

```python
import matplotlib
cmap= matplotlib.colors.LinearSegmentedColormap.from_list("",["#D80F0F","#191313"])

gorsel= WordCloud(background_color= "white",
                 width=500,
                 height=400,
                 max_words= 10,
                 colormap= cmap).generate(text)


plt.figure(figsize=(10,10))
plt.imshow(gorsel)
plt.axis("off")
plt.show()
```
