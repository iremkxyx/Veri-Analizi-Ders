```python
import numpy as np
import pandas as pd
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv(r'C:\Users\Lenovo\Data Analysis\Projeler\verisetleri\users.data', sep='\t', names=column_names)
```

```python
df.head()
```

```python
len(df) #kayıt sayısı
```

```python

movie_titles = pd.read_csv("movie_id_titles.csv")
movie_titles.head()
```

```python
len(movie_titles)
```

```python

df = pd.merge(df, movie_titles, on='item_id')
df.head()
```

```python

moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
type(moviemat)
```

```python
# Amaç: Starwars filmine benzer film önerileri yapmak
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
starwars_user_ratings.head()
```

```python
# corrwith() metodunu kullanarak Star wars filmi ile korelasyonları hesaplatalım:
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_starwars
```

```python
type(similar_to_starwars)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
#  alakasız sonuçlar çıktı, bu konuyu biraz araştırınca bunun nedeninin bu filmlerin çok az oy alması nedeniyle olduğunu bulacaksınız.. Bu durumu düzeltmek için 100'den az oy alan filmleri eleyelim.. Bu amaçla ratings isimli bir dataframe oluşturalım ve burada her fimin kaç tane oy aldığını (yani oy sayısını) tutalım...
df.head()
```

```python
df.drop(['timestamp'], axis = 1)
```

```python
# her filmin ortalama rating değerini bulalım 
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())

ratings.sort_values('rating',ascending=False).head()
```

```python
# bu ortalamalar hesaplanırken kaç oy aldığına bakmadık o yüzden böyle hiç bilmediğimiz filmler çıktı
# her filmin aldığı oy sayısını bulalım..
ratings['rating_oy_sayisi'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
# en çok oy alan filmleri büyükten küçüğe sıralayalım
ratings.sort_values('rating_oy_sayisi',ascending=False).head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['rating_oy_sayisi'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['rating_oy_sayisi']>500].sort_values('Correlation',ascending=False).head()
```
