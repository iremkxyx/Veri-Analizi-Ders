```python
import numpy as np
import pandas as pd
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv(r'C:\Users\Lenovo\Data Analysis\Projeler\verisetleri\users.data', sep='\t', names=column_names)
```

```python

movie_titles = pd.read_csv("movie_id_titles.csv")
movie_titles.head()
```

```python
df = pd.merge(df, movie_titles, on='item_id')
df.head()
```

```python

moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
toystory_user_ratings = moviemat['Toy Story (1995)']
toystory_user_ratings.head()
```

```python
similar_to_toystory = moviemat.corrwith(toystory_user_ratings)
```

```python
similar_to_toystory
```

```python
corr_toystory = pd.DataFrame(similar_to_toystory, columns=['Correlation'])
corr_toystory.dropna(inplace=True)
corr_toystory.head()
```

```python
corr_toystory.sort_values('Correlation',ascending=False).head(10)
```

```python
df.drop(['timestamp'], axis = 1)
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())

ratings.sort_values('rating',ascending=False).head()
```

```python
ratings['rating_oy_sayisi'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
ratings.sort_values('rating_oy_sayisi',ascending=False).head()
```

```python
corr_toystory.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_toystory = corr_toystory.join(ratings['rating_oy_sayisi'])
corr_toystory.head()
```

```python
corr_toystory[corr_toystory['rating_oy_sayisi']>100].sort_values('Correlation',ascending=False).head()
```
